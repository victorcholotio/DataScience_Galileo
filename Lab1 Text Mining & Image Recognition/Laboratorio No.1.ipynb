# Problema 1

Desarrolle una funcion la cual reciba dos parametros, una imagen y un entero llamado color, la funcion debe devolver una imagen la cual tenga activos los canales de color segun los siguientes puntos:  
- Si el parametro color vale 1, la imagen debe mostrar activos unicamente el color azul.  
- Si el parametro color vale 2, la imagen debe mostrar activos unicamente el color verde.  
- Si el parametro color vale 3, la imagen debe mostrar activos unicamente el color rojo.  
- Si el parametro color vale 10, la imagen debe mostrar activos unicamente los colores rojo y verde.  
- Si el parametro color vale 20, la imagen debe mostrar activos unicamente los colores verde y azul.  
- Si el parametro color vale 30, la imagen debe mostrar activos unicamente los colores azul y rojo.  

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

#Funcion para cambiar el canal de color dependiendo del nuemero ingresado
def canales(imagen, color):  
    img = cv2.imread(imagen) # Cargar la imagen
    nuevos_canales = np.zeros_like(img) #Para crear un lienzo con las mismas medidas que la imagen original

    if color == 1: # Para activar el color azul
        nuevos_canales[:, :, 0] = img[:, :, 0]
    elif color == 2: # Para activar el color  verde
        nuevos_canales[:, :, 1] = img[:, :, 1]
    elif color == 3: # Para activar el color rojo
        nuevos_canales[:, :, 2] = img[:, :, 2]
    elif color == 10: # Para activar el color rojo y verde
        nuevos_canales[:, :, 1] = img[:, :, 1]
        nuevos_canales[:, :, 2] = img[:, :, 2]
    elif color == 20:  # Para activar el color verde y azul
        nuevos_canales[:, :, 0] = img[:, :, 0]
        nuevos_canales[:, :, 1] = img[:, :, 1]
    elif color == 30:  # Para activar el color azul y rojo
        nuevos_canales[:, :, 0] = img[:, :, 0]
        nuevos_canales[:, :, 2] = img[:, :, 2]

    return nuevos_canales
```

```python
imagen = r'D:\MAESTRIA CIENCIA DE DATOS\8. TEXT MINING & IMAGE RECOGNITION\3. Laboratorios\Lab1\ny.jpg'
color = 10 #Para rojo y verde
prueba = canales(imagen,color)
cv2.imwrite('D:\\MAESTRIA CIENCIA DE DATOS\\8. TEXT MINING & IMAGE RECOGNITION\\3. Laboratorios\\Lab1\\ny_modificada.jpg', prueba)
```

```python
prueba = cv2.cvtColor(prueba, cv2.COLOR_BGR2RGB) #conversión a estructura correcta de RGB
plt.imshow(prueba)
plt.show()
```

# Problema 2

En el .zip del laboratorio se le compartio un conjunto de imagenes en escala de grises (imagen1,imagen2, perro) estas imagenes fueron creadas utilizando una escala de grises en 3D, cree una funcion que dadas las 3 imagenes se construya la imagen original a color.

```python
def reconstruccion(imagen_azul, imagen_verde, imagen_rojo):
    
    # Leer las imagenes
    img_azul = cv2.imread(imagen_azul, cv2.IMREAD_GRAYSCALE)
    img_verde = cv2.imread(imagen_verde, cv2.IMREAD_GRAYSCALE)
    img_rojo = cv2.imread(imagen_rojo, cv2.IMREAD_GRAYSCALE)

    imagen_color = cv2.merge([img_azul, img_verde, img_rojo]) # Merge para unir las imagenes

    return imagen_color
```

```python
def generar_grises(imagen_color, img_salida_azul, img_salida_verde, img_salida_rojo):
    #Leer la imagen
    img_color = cv2.imread(imagen_color)

    #Extraemos los distintos canales para la imagen
    img_azul, img_verde, img_rojo = cv2.split(img_color)
    
    #Guardar cada uno de los canales
    cv2.imwrite(img_salida_azul, img_azul)
    cv2.imwrite(img_salida_verde, img_verde)
    cv2.imwrite(img_salida_rojo, img_rojo)
```

# Problema 4

Cree una funcion que dada una imagen, muestre el histograma de cada canal de color y el de escala de grises (utilice un promedio aritmetico para su escala de grises, no puede usar funciones de opencv), sus histogramas deben incluir una linea vertical la cual muestre el valor de la media de la distribucion.

```python
def histograma(imagen):
    hist, bins = np.histogram(imagen.flatten(), bins = 256, range =[0, 256])
    return hist

def print_histograma(imagen_color):
    
    img_color = cv2.imread(imagen_color) 
    canales = cv2.split(img_color)
    colores = ('b','g', 'r')

    plt.figure(figsize=(12,8))
    for i, (canal,color) in enumerate(zip(canales, colores)):
        hist = histograma(canal)
        valor_promedio = np.mean(canal)

        plt.subplot(2, 2, i+1)
        plt.plot(hist, color=color)
        plt.axvline(x=valor_promedio, color='k', linestyle='dashed', linewidth=1)
        plt.title(f'Histograma del canal {color.upper()}')
        plt.xlabel('Intensidad')
        plt.ylabel('Frecuencia')

    #imagen escala de grises usando promedio y calculando el histograma
    imagen_gris = np.mean(img_color, axis=2).astype(np.uint8)

    histograma_gris = histograma(imagen_gris)
    promedio_escala_gris = np.mean(imagen_gris)

    plt.subplot(2, 2, 4)
    plt.plot(histograma_gris, color='k')
    plt.axvline(x=promedio_escala_gris, color='r', linestyle='dashed', linewidth=1)
    plt.title('Histograma de la imagen en escala de grises')
    plt.xlabel('Intensidad')
    plt.ylabel('Frecuencia')

    plt.tight_layout()
    plt.show()    
```

```python
ny = r'D:\MAESTRIA CIENCIA DE DATOS\8. TEXT MINING & IMAGE RECOGNITION\3. Laboratorios\Lab1\ny.jpg'
print_histograma(ny)
```

# Problema 5

Investigue en que consiste el enfoque de escala de grises ponderado, luego de esto implemente una funcion que dada una imagen, realice una escala de grises ponderada (notar que no existe una solucion unica).

```python
def promedio_ponderado(imagen_a_color, ponderacion=(0.299, 0.587, 0.114)):
    # formula 1 = Red*0.299 + Green*0.587 + Blue*0.114
    # formula 2 = Red*0.2126 + Green*0.7152 + Blue*0.0722
    img_color = cv2.imread (imagen_a_color)

    azul, verde, rojo = cv2.split(img_color)

    #Formula 1
    imagen_gris = (ponderacion[0]*rojo+ponderacion[1]*verde+ponderacion[2]*azul).astype(np.uint8)

    return imagen_gris
    
```

```python
ny = r'D:\MAESTRIA CIENCIA DE DATOS\8. TEXT MINING & IMAGE RECOGNITION\3. Laboratorios\Lab1\ny.jpg'
gris_ponderado = promedio_ponderado(ny)

cv2.imwrite(r'D:\MAESTRIA CIENCIA DE DATOS\8. TEXT MINING & IMAGE RECOGNITION\3. Laboratorios\Lab1\imagen_gris_ponderado.jpg', gris_ponderado)
```

### El modelo de color HSV
El modelo de color HSV se basa en un sistema de color en el que el espacio de color se representa mediante un único cono. Los tres componentes del cono son tono, saturación y valor, como se muestra en la siguiente ilustración:

  ![Imagen color HSV](hsv.png)

Los colores enel modelo HSV se mapea de la siguiente manera

Hue (Tono):

- Representa el color en sí mismo.
- Es el ángulo en el círculo de colores y va desde 0° a 360°.
- Por ejemplo, 0° es rojo, 120° es verde, y 240° es azul.

Saturation (Saturación):

- Representa la pureza o intensidad del color.
- Va desde 0 a 100%.
- 0% es un tono gris y 100% es el color más vibrante (sin mezcla de blanco).


Value (Valor):

- Representa el brillo del color.
- Va desde 0 a 100%.
- 0% es completamente oscuro (negro) y 100% es el color en su máxima luminosidad.
